```python
# 1. Подключение библиотеки numpy
import numpy as np

# 1.1. Создание одномерного массива my_array, в котором по порядку все числа до 100, кратные 5
my_array = np.arange(5, 101, 5)
print("my_array:", my_array)

# 2.1. Получение массива с кратными 10 числами
array_div_10 = my_array[my_array % 10 == 0]
print("array_div_10:", array_div_10)

# 3.1. Получение двумерного массива с двумя столбцами
two_col_array = my_array.reshape(-1, 2)
print("two_col_array:\n", two_col_array)

# 4.1. Получение массива my_second_array, где каждый элемент на 5 меньше исходного
my_second_array = my_array - 5
print("my_second_array:", my_second_array)
# Повторим строки в two_col_array, чтобы число строк совпадало с my_second_array
repeated_two_col_array = np.repeat(two_col_array, 2, axis=0)

# Теперь объединяем массивы
combined_array = np.hstack((repeated_two_col_array, my_second_array.reshape(-1, 1)))
print("combined_array:\n", combined_array)

# 6.1. Разделение 2 и 3 строки на 4 верхних строки my_array. Использование .T для транспонирования
div_array = (combined_array[1:3, :] / combined_array[0:4, :].T)
print("div_array:\n", div_array)

# 7.1. Оставляем только те значения, которые не изменились после деления
unchanged_values = div_array[np.isclose(div_array, 1)]
print("unchanged_values:", unchanged_values)

# 8.1. Снова получаем трехстрочный двумерный массив и меняем местами первую и третью строки
three_row_array = combined_array[:3, :]
three_row_array[[0, 2]] = three_row_array[[2, 0]]
print("three_row_array:\n", three_row_array)

# 9.1. Вычисление среднего, медианы и стандартного отклонения для каждой строки
mean_values = np.mean(three_row_array, axis=1)
median_values = np.median(three_row_array, axis=1)
std_values = np.std(three_row_array, axis=1)

print("Средние значения по строкам:", mean_values)
print("Медианы по строкам:", median_values)
print("Стандартные отклонения по строкам:", std_values)

# 10.1. Нормализация на максимальное значение вдоль строк
normalized_array = three_row_array / np.max(three_row_array, axis=1, keepdims=True)
print("Нормализованный массив:\n", normalized_array)
```
